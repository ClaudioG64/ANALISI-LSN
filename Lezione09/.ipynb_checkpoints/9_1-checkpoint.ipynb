Il Traveling Salesman Problem (TSP) è un problema di ottimizzazione combinatoria classico in cui si cerca di trovare il percorso più breve che permetta a un venditore di visitare un insieme di città una sola volta e tornare alla città di partenza.

```python
import matplotlib.pyplot as plt

# Leggi i dati dal file
x_coords = []
y_coords = []
with open('cities_circle.dat', 'r') as file:
    for line in file:
        data = line.strip().split()
        x_coords.append(float(data[0]))
        y_coords.append(float(data[1]))

# Crea il plot
plt.figure(figsize=(5, 5))
plt.scatter(x_coords, y_coords, color='blue', s=50)

# Aggiungi etichette numerate
for i, (x, y) in enumerate(zip(x_coords, y_coords), start=1):
    plt.annotate(str(i), (x, y), xytext=(5, 5), 
                 textcoords='offset points', 
                 fontsize=8,
                 alpha=0.7)

plt.xlabel('Coordinata X')
plt.ylabel('Coordinata Y')
plt.title('Cities (circumference)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Analisi performance:

$L^{(2)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|^2$

$L^{(1)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|$

Nel nostro algoritmo genetico, la fitness rappresenta la qualità di una soluzione (percorso) ed è definita come l'inverso del costo del percorso.

L'andamento della fitness attraverso le generazioni (visualizzato nei grafici) mostra come l'algoritmo migliora progressivamente la qualità della soluzione, avvicinandosi al percorso ottimale.

NB: i dati relativi alla fitness media si riferiscono solo alla metà migliore degli individui di ogni generazione.

```python
import matplotlib.pyplot as plt
import numpy as np

# Inizializza le liste per i dati
generations = []
best_fitness = []
average_fitness = []

# Leggi i dati dal file
with open('fitness_circle.dat', 'r') as file:
    for line in file:
        data = line.strip().split()
        if len(data) >= 3:
            generations.append(int(data[0]))
            best_fitness.append(float(data[1]))
            average_fitness.append(float(data[2]))

# Crea il plot
plt.figure(figsize=(10, 6))

# Plot della miglior fitness
plt.plot(generations, best_fitness, 'b-', linewidth=2, label='Migliore fitness')
plt.plot(generations, best_fitness, 'bo', markersize=4)

# Plot della fitness media
plt.plot(generations, average_fitness, 'r-', linewidth=2, label='Fitness media')
plt.plot(generations, average_fitness, 'ro', markersize=4)

plt.xlabel('Generazione')
plt.ylabel('Fitness')
plt.title('Performance Algoritmo Genetico')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()

# Aggiungi annotazioni per i punti significativi (opzionale)
# Trova il punto con la miglior fitness
best_idx = np.argmin(best_fitness)  # Usa argmax se valori più alti sono migliori
plt.annotate(f'Migliore: {best_fitness[best_idx]:.2f}', 
             xy=(generations[best_idx], best_fitness[best_idx]),
             xytext=(10, 10), textcoords='offset points',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))

plt.show()

# Opzionale: stampa alcune statistiche
print(f"Numero di generazioni: {len(generations)}")
print(f"Miglior fitness raggiunta: {min(best_fitness):.4f}")  # Usa max() se valori alti sono migliori
print(f"Fitness media finale: {average_fitness[-1]:.4f}")
```
