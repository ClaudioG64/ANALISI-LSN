```python
import matplotlib.pyplot as plt
import numpy as np

# Carica le posizioni delle città
cities = np.loadtxt('cities_ita.dat')
x_cities = cities[:, 0]
y_cities = cities[:, 1]

# Carica il percorso ottimale
path = np.loadtxt('global_best_path_ita.dat')
x_path = path[:, 0]
y_path = path[:, 1]

# Crea il grafico
plt.figure(figsize=(10, 8))
plt.plot(x_cities, y_cities, 'o', markersize=8, label='Città', alpha=0.7)
plt.plot(x_path, y_path, 'r-', linewidth=2, label='Percorso ottimale')
plt.plot([x_path[-1], x_path[0]], [y_path[-1], y_path[0]], 'r-', linewidth=2)  # Chiudi il percorso

# Aggiungi etichette e titolo
plt.xlabel('Coordinate X')
plt.ylabel('Coordinate Y')
plt.title('Percorso Ottimale TSP - Città italiane')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('equal')  # Mantiene le proporzioni corrette

# Salva e mostra il grafico
plt.savefig('tsp_ita_result.png', dpi=300, bbox_inches='tight')
plt.show()

# Calcola alcune statistiche
path_length = 0
for i in range(len(path)-1):
    dx = path[i+1, 0] - path[i, 0]
    dy = path[i+1, 1] - path[i, 1]
    path_length += np.sqrt(dx*dx + dy*dy)

print(f"Lunghezza totale del percorso: {path_length:.4f}")
print(f"Numero di città: {len(cities)}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Carica i dati della fitness
data = np.loadtxt('fitness_history_ita.dat')
generations = data[:, 0]
best_fitness = data[:, 1]
avg_fitness = data[:, 2]

# Crea il grafico
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(generations, best_fitness, 'b-', label='Best Fitness')
plt.plot(generations, avg_fitness, 'r--', label='Average Fitness (best half)')
plt.xlabel('Generations')
plt.ylabel('Fitness')
plt.title('TSP Performance - Italian Cities')
plt.legend()
plt.grid(True)

# Carica il percorso ottimale
cities = np.loadtxt('cities_ita.dat')
path = np.loadtxt('global_best_path_ita.dat')

plt.subplot(1, 2, 2)
plt.plot(cities[:, 0], cities[:, 1], 'o', markersize=4, label='Cities')
plt.plot(path[:, 0], path[:, 1], 'r-', linewidth=1, label='Optimal Path')
plt.plot([path[-1, 0], path[0, 0]], [path[-1, 1], path[0, 1]], 'r-', linewidth=1)
plt.xlabel('X (normalized)')
plt.ylabel('Y (normalized)')
plt.title('Optimal TSP Path - Italian Cities')
plt.legend()
plt.grid(True)
plt.axis('equal')

plt.tight_layout()
plt.savefig('tsp_italian_performance.png', dpi=300)
plt.show()

# Calcola statistiche
total_distance = 0
for i in range(len(path)-1):
    dx = path[i+1, 0] - path[i, 0]
    dy = path[i+1, 1] - path[i, 1]
    total_distance += np.sqrt(dx*dx + dy*dy)

print(f"Total path length: {total_distance:.4f}")
print(f"Number of cities: {len(cities)}")
print(f"Final best fitness: {best_fitness[-1]:.4f}")
print(f"Improvement from initial: {((best_fitness[0] - best_fitness[-1]) / best_fitness[0] * 100):.2f}%")
```
