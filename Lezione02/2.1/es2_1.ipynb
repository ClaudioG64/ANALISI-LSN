```python
import numpy as np
import matplotlib.pyplot as plt

# Carica i dati da output.dat
data = np.loadtxt("output.dat", skiprows=1)  # Salta la prima riga con l'intestazione

# Estrai le colonne
N = data[:, 0]  # Numero di blocchi inclusi progressivamente
mean_uniform = data[:, 1]
error_uniform = data[:, 2]
mean_importance = data[:, 3]
error_importance = data[:, 4]

# Valore atteso dell'integrale (da calcolo analitico)
I_exact = 1.0

# Grafico per il metodo uniforme
plt.figure(figsize=(10,5))
plt.errorbar(N, mean_uniform, yerr=error_uniform, fmt="o-", label="Uniform Sampling")
plt.axhline(I_exact, color="r", linestyle="dashed", label="Valore esatto")
plt.xlabel("Numero di blocchi")
plt.ylabel("Stima dell'integrale")
plt.title("Convergenza del metodo Monte Carlo - Uniform Sampling")
plt.legend()
plt.grid()
plt.show()

# Grafico per il metodo con importance sampling
plt.figure(figsize=(10,5))
plt.errorbar(N, mean_importance, yerr=error_importance, fmt="o-", label="Importance Sampling")
plt.axhline(I_exact, color="r", linestyle="dashed", label="Valore esatto")
plt.xlabel("Numero di blocchi")
plt.ylabel("Stima dell'integrale")
plt.title("Convergenza del metodo Monte Carlo - Importance Sampling")
plt.legend()
plt.grid()
plt.show()
```
