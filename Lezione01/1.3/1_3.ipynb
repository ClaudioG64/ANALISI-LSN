```python
import numpy as np
import matplotlib.pyplot as plt

# Carico i dati dal file
chi2_values = np.loadtxt("output1_3.dat", usecols=1)

# Parametri
N = 100  # numero di intervalli usati nel C++ (da modificare se cambi nel codice)

# Plot istogramma
plt.figure(figsize=(8,5))
plt.hist(chi2_values, bins=20, edgecolor='black', alpha=0.7)
plt.axvline(N, color='red', linestyle='--', label=f"Valore atteso ~ {N}")
plt.xlabel(r"$\chi^2$")
plt.ylabel("Frequenza")
plt.title(r"Distribuzione dei valori di $\chi^2$")
plt.legend()
plt.grid(True)
plt.show()
```
