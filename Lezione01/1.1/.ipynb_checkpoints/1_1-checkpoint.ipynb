```python
import numpy as np
import matplotlib.pyplot as plt

# Carico i dati
data = np.loadtxt("output1_1.dat")

blocks = data[:,0]     # Numero di blocchi
mean = data[:,1]       # Media progressiva
error = data[:,2]      # Errore statistico

# Valore teorico atteso = 0.5
theory = 0.5

# Plot
plt.figure(figsize=(8,5))
plt.errorbar(blocks, mean, yerr=error, fmt='o', capsize=3, label="Data Blocking")
plt.axhline(theory, color='red', linestyle='--', label="Valore teorico 0.5")
plt.xlabel("Numero di blocchi")
plt.ylabel("Media dei numeri casuali")
plt.title("Stima di <r> con Data Blocking")
plt.legend()
plt.grid(True)
plt.show()
```
